```python
### Regular expression module -> One of the most used library

import re

pattern = r'\d+'
text = "I got 100 marks in coding"

match = re.search(pattern,text)
print(match)
print(match.group())
```

```python
### Date module in python

from datetime import datetime, timedelta ### timedelta is used for yesterdays data

result = datetime.today()
print(result)

result2 = datetime.now()
print(result2)

yesterday = result - timedelta(days=1)
print(yesterday)
```

```python
### creating a CSV file and writing to the file and reading the file

import csv

with open('example.csv', mode='w',newline='') as file:
    writter = csv.writer(file)
    writter.writerow(['name','id'])
    writter.writerow(['Rahul','32'])

with open('example.csv',mode='r') as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)
```

```python
### module json , to get data in json format

import json

data = {
    'name' : 'Rahul',
    'age'  : 25,
    'meaningOfLife' : 'dontKnow'
}

dictToJson = json.dumps(data)
print(dictToJson)
print(type(dictToJson))

JsonToDict = json.loads(dictToJson)
print(JsonToDict)
print(type(JsonToDict))

```
