```python
### create a CSV

import csv

with open('example2.csv',mode='w',newline='') as file:
    writter = csv.writer(file)
    writter.writerow(['name','age','interest'])
    writter.writerow(['Rahul','25','Space'])

with open('example2.csv',mode='r') as file:
    reader = csv.reader(file)

    for line in reader:
        print(line)
```

```python
### Python data to json format

import json

data = {
    'Rahul' : {
        'nickName' : 'Bunny',
        'age'  : 25,
    },
    'Sushma' : {
        'nickName' : 'chinni',
        'age'      : 33
    }
}

dataToJson = json.dumps(data)
print(dataToJson)
print(type(dataToJson))

JsonToData = json.loads(dataToJson)
print(JsonToData)
print(type(JsonToData))
```

```python
### Use the `random` module to generate a list of 5 random 
# numbers between 1 and 50 and shuffle the elements of a list.

import random

print(random.randrange(1,50))

numbers = random.sample(range(1,50),5)
print(numbers)

random.shuffle(numbers)
print(numbers)


```

```python
### Use the `datetime` module to print the current date, calculate the date 
# 100 days from today, and determine the day of the week for a given date.

from datetime import datetime, timedelta

today = datetime.now()
print(f"current date is {today}")

hundardDaysBack = today - timedelta(100)
print(f"100 days back the date is {hundardDaysBack}")

print(hundardDaysBack.strftime("%A"))  ### %A gives full week name
                                       ### %a gives half week name
                                       ### %y gives year
                                       ### %m gives month
                                       ### %d gives day


```
