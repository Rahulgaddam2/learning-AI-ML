```python
### Memoization means -> Remember the answer so you don’t calculate it again.
### But how do you remember in python?
### -> We use a dictionary.
### Dictionary is like : question → answer
### Memoization = dictionary + remember results + skip repeated work

nano = {}  ## declare a dictionary
def square(n):
    if n in nano:
        return nano[n]
    
    result = n*n
    nano[n] = result
    print(result) 

square(4)
```

```python
### Now lets do recursion with Memoization
### -> Memoization is added to recursion to remember results of previous recursive calls.
### Here we have 5 steps:
### 1 ->  If result already remembered → return it
### 2 ->  If base case → return answer
### 3 ->  Otherwise → do recursion
### 4 ->  Save result
### 5 ->  Return result
```

```python
### Problem: Define a recursive function to calculate the nth Fibonacci number using memoization.

namo = {}

def recursionWithMemoization(number):

    if number in namo:
        return namo[number]
    
    if number == 0:
        return 1
    
    namo[number] =  number* recursionWithMemoization(number-1)
    return namo[number]

recursionWithMemoization(5)
```

```python
### Define a recursive function to calculate the nth Fibonacci number using memoization. Test the function with different inputs.

memo = {}
def Fibonacci(number):

    if number in memo:
        return number
    
    if number == 0:
        return 0
    
    if number == 1:
        return 1

    memo[number] = Fibonacci(number-1) + Fibonacci(number-2)
    return memo[number]

Fibonacci(5)
```
