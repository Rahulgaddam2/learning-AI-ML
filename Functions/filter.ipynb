```python
### finding numbers which are greater than 5 without using lambda function

listOfNumbers = [1,88,3,55,77,3,5,67,7,9]

def greaterThanFive(Numbers):
    if Numbers > 5:
        return True

result1 = list(filter(greaterThanFive,listOfNumbers))
print(result1)
```

```python
### The filter function is used to filter out the items from an list or any iterable bases on a condition

listss = [1,2,3,4,5,6,7,8,9,0,10]

result = list(filter(lambda x : x%2==0, listss))
print(result)
```
