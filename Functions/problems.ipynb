```python
### A Function to read a file and count the frequency of each word

def fileFrequncyCount(file_path):
    wordCount = {}
    with open(file_path, 'r') as file:
        for line in file:
            words =  line.split()
            for word in words:
                word = word.lower().strip('.,!?;:"\'')
                wordCount[word] = wordCount.get(word, 0) +1

    return wordCount

file_path = 'fileExample.txt'
word_frequncy = fileFrequncyCount(file_path)
print(word_frequncy)
```

```python
### Calculate the factorials of a number using recursion

def factorialRecursion(Number):

    if Number == 0:
        return 1
    

    return Number * factorialRecursion(Number-1)


result = factorialRecursion(5)
print(result)
```

```python
### Check if the string is palindrom
### ABA is a palindrom

def checkPalindrom(value):

    value = value.replace(" ","")
    start = 0
    end = len(value) -1

    while start < end:
        
        if value[start] != value[end]:
            print("Its not a palindrom")
            return
        start += 1
        end -= 1

    print("Its a palindrom")

checkPalindrom("rhtthrr")
```

```python
### calculate the total cost of items in a shopping cart
### example of cart data



cart = [
    {'name':'kit','prize':44.5,'quantity':4},
    {'name':'ball','prize':3.0,'quantity':10},
    {'name':'bat','prize':7.5,'quantity':10}
]

def shoppingCartTotalPrize(cart):
    totalCost = 0
    for item in cart:
        value = item['prize'] * item['quantity']
        totalCost += value

    return totalCost
        

shoppingCartTotalPrize(cart)

```

```python
### Temperature conversion

def conversion(temp,unit):
    ### This function converts temperature between celsius and fahrenheit
    if unit=='C':
        result = temp* 9/5 + 32  ### celsius to fahrenheit
        return result
    elif unit=='F':
        result = (temp-32)*5/9  ### fahenheit to calsius
        return result
    else:
        print("please enter the valid temp and unit")

result = conversion(888,'C')
print(result)
result2 = conversion(1630.4,'F')
print(result2)
```
