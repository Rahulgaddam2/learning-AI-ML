```python
### Return statement

def returningValue(num1,num2):
    return num1+num2

result = returningValue(2,2)
print(result)

### In python return can return multipul values

def returningMultiValues(num1,num2):
    return num1+num2, num1, num2

result = returningMultiValues(3,3)
print(result)
```

```python
### Both positional arguments and keywords arguments in one function

def combine(*args, **kwargs):
    for val in args:
        print(f"positional argument = {val}")
    
    for key,value in kwargs.items():
        print(f"{key} = {value}")

combine(1,2,3,"Sushma","Adhvay",name="rahul",age=26) ### Always give positional arguments first than keyword arguments
```

```python
### keyword arguments
def printDetails(**kwargs): ### All the parameters with be under key-value pair
    for key,value in kwargs.items():
        print(f"{key},{value}")

printDetails(name="Rahul", age="32", country="USA")

```

```python
### Positional arguments

def printNumbers(*rahul):  ## when you have multipul parameters use * , in place of rahul use args for standard
    for number in rahul:
        print(number)

printNumbers(1,2,3,"Rahul",4,5,"Sushma") ## here you are giving multipul parameters, these parameters are called as positional parameters

```

```python
### Default parameters

def greet(name):
    print(f"Welcome to my world {name}")

greet("rahul")


### now you can keep an default value in place of name, so if you forget the parameter the default value will get as parameter

def greetWithDefault(name="Default"):
    print(f"welcome to the world {name}")

greetWithDefault() ### here the parameter will we default value which we have given in the function
```

```python
### Functions with parameters

def addingTwoNumbers(number1,number2):
    return number1+number2

result = addingTwoNumbers(2,2)
print(result)
```
