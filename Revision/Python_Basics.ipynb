```python
### Question 3:Write a Python program to check if a number is positive, negative, or zero.

def checkNumber(number):
    if number > 0:
        print(f"The number: {number} is a positive number")
    elif number < 0:
        print(f"The number: {number} is a negative number")
    else:
        print(f"The number: {number} is Zero")

checkNumber(3)
```

```python
### Question 4: Write a Python program to find the largest of three numbers.

def findingLargeNumber(num1,num2,num3):

    if num1 > num2 and num1 > num3:
        print(f"The largest number is {num1}")
    elif num2 > num1 and num2 > num3:
        print(f"The largest number is {num2}")
    else:
        print(f"The largest number is {num3}")

findingLargeNumber(11,22,3)

### We can also use max inBuilt function, max can take any number of parameters

def findingLargeNumberUsingMax(num1,num2,num3):

    result = max(num1,num2,num3)
    
    return result

answer = findingLargeNumberUsingMax(11,22,3)
print(answer)
```

```python
### Question 5: Write a Python program to calculate the factorial of a number.

def findingFactorial(number):

    if number == 0:
        return 1
    
    return number*findingFactorial(number-1)

value = 5
findingFactorial(value)
```

```python
### Question 7: Write a Python program to swap the values of two variables.

def swapper(value1,value2):

    temp = value1
    value1 = value2
    value2 = temp

    return value1,value2

swapper(5,10)
```

```python
### Question 16: Write a Python program to find the sum of the first n natural numbers.

def firstNNaturalNumbers(value):
    sum = 0
    while value > 0:
        sum += value
        value -= 1
    
    return sum

firstNNaturalNumbers(10)
```

```python
### Question 17: Write a Python program to check if a year is a leap year.

def leapYear(year):

    if (year%400 == 0) or (year% 4 == 0 and year% 100 != 0):
        print("Its a leap year") 
    else:
        print("Its not a leap year")

leapYear(2016)
```

```python
### Question 19: Write a Python program to check if a string is a palindrome.

def palindrome(value):

    start = 0
    end = len(value) -1

    while start < end:
        if value[start] == value[end]:
            start += 1
            end -= 1
        else:
            print("Its not a palindrome")
    
    print("Its a palindrome")

palindrome("abccba")
```

```python
### Question 20: Write a Python program to sort a list of numbers in ascending order.

def sorting(list):

    return list.sort()

sorted([1,7,4,8,2,6])
```

```python
### Question 20: Write a Python program to sort a list of numbers in descending order.

def descending(lst):
    lst.sort(reverse=True)
    return lst

print(descending([1,7,4,8,2,6]))

```
