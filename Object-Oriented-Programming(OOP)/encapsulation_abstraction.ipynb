```python
### encapsulation and abstraction

### encapsulation is the concept of wrapping data(variables) and methods(functions)
### together as a single unit 
### It restricts direct access to some of the object's components

### Encapsulation is done with getter and setter methods

### There can be 3 types of variables
### / public variable / private variable / protected variable
```

```python
### public variables

class Student:
    def __init__(self, name, age):
        self.name = name   ### public variable
        self.age = age     ### public variable

    def nameStudent(self):
        print(f"name of the student is : {self.name}")

student1 = Student("Rahul", 25)
## dir(student1)  ### The public variable can be accessed from outside

### example

def nameOfStudent(student):
    student.nameStudent()   ### we can access the variable from outside the class

nameOfStudent(student1)
```

```python
### private variables
### use 2 underscores to create a private variable
class Student:
    def __init__(self, name, age):
        self.__name = name   ### private variable 
        self.__age = age     ### private variable

    def nameStudent(self):
        print(f"name of the student is : {self.name}")

student1 = Student("Rahul", 25)
## dir(student1)  ### The private variable can not be accessed from outside 

### example

def nameOfStudent(student):
    student.nameStudent()   ### we can not access the variable from outside the class

nameOfStudent(student1)  ### you will get an error here AttributeError: 'Student' object has no attribute 'name'
```

```python
### protected Variable
### We can only use the variables when we inherti the variable class
### use single underscore to create a protected variable

class Student:
    def __init__(self, name, age):
        self._name = name   ### protected variable 
        self._age = age     ### protected variable

    def nameStudent(self):
        print(f"name of the student is : {self.name}")

class inheriting(Student):
    def __init__(self,name,age):
        super().__init__(name,age)

    def details(self):
        print(f"name of the student is : {self._name}")
        print(f"age of the student is : {self._age}")


stu = inheriting("Sushma",33)

stu.details()
```

```python
### Encapsulation with getter and setter

class Person:
    def __init__(self,name,age):
        self.__name = name
        self.__age = age
    
    def get_name(self):
        print(f"name of the person is : {self.__name}")
    
    def set_name(self,name):
        self.__name = name
    
    def get_age(self):
        print(f"The age of the preson is : {self.__age}")

    def set_age(self,age):

        if age > 0:
            self.__age = age
        else:
            print("Age can not be negative")

person1 = Person("Rahul",25)

person1.get_name()
person1.get_age()

### now I am setting the values which will replace the old values

person1.set_name("Sushma")
person1.set_age(33)

person1.get_name()
person1.get_age()
```

```python
### Now lets discuss about Abstraction

### Abstraction

### Abstraction is concept of hiding the complex implementation details and showing
### only the necessary features of an object
```

```python
from abc import ABC,abstractmethod

### abstract class

class Vehicle(ABC):
    def drive(self):  ### This is an normal method
        print("The vehicle is used for driving")

    @abstractmethod
    def start_engine(self):
        pass


class Car(Vehicle):
    def start_engine(self):
        print("Car enginer stated")


def operate_vehicle(vehcile):
    vehcile.start_engine()

car = Car()
operate_vehicle(car)
```
