```python
### What is Operator Overloading?
### Giving a new meaning to an operator (+, -, *, ==, etc.) when it is used with objects (classes).

### Normally, operators work with numbers: print(5 + 3)   # 8
### But with operator overloading, we can decide how an operator behaves for our own class.
```

```python
### + → __add__()
### - → __sub__()
### * → __mul__()
### / → __truediv__()
### == → __eq__()
### < → __lt__()
### > → __gt__() , and more
```

```python
class Vector:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    
    def __add__(self,other):
        return Vector(self.x + other.x, self.y + other.y)
    
    def __sub__(self,other):
        return Vector(self.x - other.x, self.y - other.y)
    
    def __mul__(self,other):
        return Vector(self.x * other.x, self.y * other.y)
    

    
    def __str__(self):
        return f"{self.x,self.y}"
    
test = Vector(2,2)
test2 = Vector(4,4)

print(test + test2)
print(test - test2)
print(test * test2)
```
