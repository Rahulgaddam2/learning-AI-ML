```python
### Polymorphsim is a core concept of oops that allows objects of different classes
### to be treated as objects of a common superclass
### It provides a way to perform a single action in different forms

### Polymorphsim is typically acheived through method overriding and interfaces

### Poly = multipul , morphsim = forms
```

```python
### What is method overriding

### Method overriding allows a child class to provide a specific implementation of a 
### method that is already defined in its parent class

```

```python
### polymorphsim with classes

class Animal:
    def speak(self):
        print("Sound of the animal")

class Dog(Animal):
    def speak(self):   ### Method overriding
        print("woof!!!")

class Cat(Animal):  
    def speak(self):   ### Method overriding
        print("meow!!!!")

## now just create a public functions

def animal_sound(animal):
    animal.speak()

dog = Dog()
dog.speak()
cat = Cat()
cat.speak()
animal_sound(dog)
```

```python
### polymorphsim with functions and methods

class shape:
    def area(self):
        return "The area of the figure"
    
class Rectangle(shape):
    def __init__(self,width,height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height
    
class Circle(shape):
    def __init__(self,radius):
        self.radius = radius
    
    def area(self):
        return 3.14*self.radius * self.radius
    
### Function to show polymorphsim

def print_area(shape):
    print(f"the are is {shape.area()}")

rectangle = Rectangle(4,5)
circle = Circle(3)

print_area(rectangle)
print_area(circle)



        
```

```python
### polymorphsim with abstract base class

from abc import ABC, abstractmethod
### base class

class Vehicle(ABC):

    @abstractmethod
    def name(self):
        pass

class Car(Vehicle):

    def name(self):
        print("The name of the vehicle is car")

class Bike(Vehicle):

    def name(self):
        print("The name of the vechile is bike")

car = Car()
car.name()

bike = Bike()
bike.name()
```
