```python
### Create a base class named `Shape` with a method `area`. 
# Create two derived classes `Circle` and `Square` that override the 
# `area` method. Create a list of `Shape` objects and call the `area` 
# method on each object to demonstrate polymorphism.

class Shape:

    def area(self):
        print("Print are of shape")

class Circle(Shape):
    def __init__(self,radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

class Square(Shape):

    def __init__(self,value):
        self.value = value

    def area(self):
        return self.value * self.value
    
listOfShapes = [Circle(5),Square(5)]

for shapes in listOfShapes:
    print(shapes.area())
```

```python
### Create a function named `describe_shape` that takes a `Shape` 
# object as an argument and calls its `area` method. Create objects of 
# `Circle` and `Square` classes and pass them to the `describe_shape` function.

class Shape:

    def test(self):
        print("This is shape class")

class Circle(Shape):
    def __init__(self,radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

class Square(Shape):

    def __init__(self,value):
        self.value = value

    def area(self):
        return self.value * self.value
    


    
def describe_shape(shape):
    print(shape.area())

circle = Circle(5)
square = Square(5)

describe_shape(circle)
describe_shape(square)
```

```python
### Create an abstract base class named `Vehicle` with an abstract method 
# `start_engine`. Create derived classes `Car` and `Bike` that implement the 
# `start_engine` method. Create objects of the derived classes and call the 
# `start_engine` method.

from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):

    def start_engine(self):
        print("Car engine started")

class Bike(Vehicle):

    def start_engine(self):
        print("Bike engine started")

def isEngineStarted(vehicle):
    return vehicle.start_engine()

car = Car()
bike = Bike()

isEngineStarted(car)
isEngineStarted(bike)
```
