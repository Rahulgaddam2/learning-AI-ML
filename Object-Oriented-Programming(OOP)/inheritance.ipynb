```python
### Inheritance is a fundamental concept in opps that allows a class to inherit attributes 
### methods from other class
### we have 2 kind of inheritance 1) Single inheritance 2) multiple inheritance
### Super keyword is used in inheritance
### super() does not call all parents; it follows MRO. In multiple inheritance it works only when all classes cooperate by using super().

```

```python
### Single inheritance

class Car:
    def __init__(self,model,windows,fuelType):
        self.model = model
        self.windows = windows
        self.fuelType = fuelType

    def fuel(self):
        print(f"uses {self.fuelType}")

    def howManyWindows(self):
        print(f"The car have : {self.windows}")


class tesla(Car):  ### here we are inheriting the Car class // passing as arguments
    def __init__(self,model,windows,fuelType,prize):
        super().__init__(model,windows,fuelType)
        self.prize = prize

    def prizeOfCar(self):
        print(f"prize of the car : {self.prize}")


car1 = tesla("XY-205",2,"electric",80000)

car1.fuel()
car1.howManyWindows()
car1.prizeOfCar()
```

```python
### multi inheritance
### Inheriting from more than one class


class Phone:
    def __init__(self,name,phone_model,color):
        self.name = name
        self.model = phone_model
        self.color = color

    def nameOfthePhone(self):
        print(f"The name of the phone : {self.name}")

    def modelOfThePhone(self):
        print(f"The model of the phone : {self.model}")

    def colorOfThePhone(self):
        print(f"The color of the phone : {self.color}")


class Computer:
    def __init__(self,computer_model,processor,ram,memory):
        self.Model = computer_model
        self.processor = processor
        self.ram = ram
        self.memory = memory

    def configurations(self):
        print(f"processor : {self.processor}")
        print(f"ram : {self.ram}")
        print(f"memory : {self.memory}")


class Student(Phone,Computer):
    def __init__(self, name, phone_model, color,computer_model,processor,ram,memory,Student_name):
        Phone.__init__(self,name,phone_model,color)
        Computer.__init__(self,computer_model,processor,ram,memory)
        self.Student_name = Student_name

    def nameOfTheStudent(self):
        print(f"name of the student is : {self.Student_name}")
            

s = Student(
    "Apple",
    "iPhone 15",
    "Black",
    "MacBook Pro",
    "M2",
    "16GB",
    "512GB",
    "Rahul"
)

s.nameOfthePhone()
s.configurations()
s.nameOfTheStudent()

        
```
