```python
### Class with instance variables and methods

class Student:
    def __init__(self,name,age):
        self.name = name
        self.age = age

### Now lets create objects

Student1 = Student("Rahul",26)
print(Student1.name)
print(Student1.age)

Student2 = Student("Sushma",33)
print(Student2.name)
```

```python
### Lets define a class with instance methods

class Student:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    
    def greet(self):
        print(f"{self.name} is from paramita high school")


student1 = Student("Rahul",25)
student1.greet()
```

```python
### Creating a bank Account class

class BankAccount:
    def __init__(self, Username, bankBalance = 0):
        self.Username = Username
        self.bankBalance = bankBalance
    
    def addingBalance(self,amount):
        self.bankBalance += amount
        print(f"{amount} is added to your bank account")

    def withdraw(self,amount):

        if amount > self.bankBalance:
            print("Insufficent balance")
        else:
            self.bankBalance -+ amount
            print(f"please collect the {amount}")
            remainingAmount = self.bankBalance - amount
            print(f"This is your remaining amount = {remainingAmount}")

    def getMyBalance(self):
        print(self.bankBalance)

account1 = BankAccount("Rahul", 5000)
account1.getMyBalance()
account1.addingBalance(3000)
account1.getMyBalance()
account1.withdraw(7000)

    
```
