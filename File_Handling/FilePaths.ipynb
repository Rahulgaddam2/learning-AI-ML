```python
### Creating a new directory
import os

new_dict = "package"
os.mkdir(new_dict)
print(f"new dict '{new_dict}' has been created")
```

```python
### listing all the files in the dict
import os

items = os.listdir('.')
print(items)
```

```python
### Joining the paths
import os

dict_name = "newFolder"
file_name = "newFile"

joinPath = os.path.join(dict_name,file_name)
print(joinPath)
```

```python
### Joining the paths with CWD
import os

dict_name = "newFolder"
file_name = "newFile"

joinPath = os.path.join(os.getcwd(),dict_name,file_name)
print(joinPath)
```

```python
### checking if the path exist or not

import os
 
fileName = 'example2.txt'

if os.path.exists(fileName):
    print(f"Yes, the file '{fileName}' exists")
else:
    print(f"There is no file with the name -> '{fileName}'")
```

```python
### Check if the path is a file or an dict
import os

fileName = 'example2.txt'

if os.path.isfile(fileName):
    print("This is a file")
elif os.path.isdir(fileName):
    print("This is a directory")
else:
    print("This is not a file or directory")
```

```python
### Getting the absolute path

import os

relative_path = "example.txt"
obsolute_path = os.path.abspath(relative_path)
print(obsolute_path)
```
