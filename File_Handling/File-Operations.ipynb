```python
### use of w+
### This mode will create a file if it is not there and overwrite the data 
### There is also r+ mode but to use this mode there should be an existing file

with open('newFile.txt',mode='w+',newline="") as file:  ## here the new file is created
    file.write("Hi, How are you\n")
    file.write("This is the new line")

    ### using seek() to point our cursor(pointer) at the start
    ### seek() is important in w+ because after writing, the 
    ###file cursor is at the END, so reading returns nothing 
    ###unless you move the cursor back.

    file.seek(0) ## 0 means first postion 

    content = file.read()
    print(content)
```

```python
### reading from an binary file

with open('example.bin',mode='rb') as binFile:
    result  = binFile.read()
    print(result)
```

```python
### Binary files
### using readbit (rb) or writebit (wb)

data = b'\x00\x01\x02\x03\x04'

with open('example.bin',mode='wb') as binFile:
    binFile.write(data)


```

```python
### writing list of lines to a file
### use function writelines

lines = ["\nThis is my first line\n","This is my second line\n","This is my 3rd line"]

with open('example.txt',mode='a') as file:
    file.writelines(lines )
```

```python
### writing a file (without overwriting)
### for this use append mode 'a'

with open('example.txt',mode='a') as file:
    file.write("\nI am using mode a -> append mode")
```

```python
### writing a file (overwriting)

with open('example.txt',mode='w',newline='') as file:
    file.write("I have overwrittern the file\n")
    file.write("this is the new line")
    ## This will remove the old text in the file and replace it with the new text
    ## Which is bad
```

```python
### Read file line by line

with open('example2.txt',mode='r') as file:
    for line in file:
        print(line.strip()) ## strip function removes the new line charecter 
```

```python
### Reading a file 

file = open('example2.txt',mode='r')

reading = file.read()
print(reading)
```

```python
### Open a file using with
with open('example.txt',mode='r') as file:
    for line in file:
        print(line)
```
