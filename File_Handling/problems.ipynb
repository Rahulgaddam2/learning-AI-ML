```python
### Read the content from a source file and write to a destination file

with open('example.txt',mode='r') as sourceFile:
    content = sourceFile.read()

with open('destination.txt',mode='w',newline='') as destinationFile:
    destinationFile.write(content)
```

```python
### Read a text file and count the number of lines, words, and charectors

with open('example.txt','r') as file:
    lines = file.readlines()
    print(len(lines))
    wordCount = sum(len(line.split()) for line in lines)
    charCount = sum(len(line) for line in lines)

    print(wordCount)
    print(charCount)

```
