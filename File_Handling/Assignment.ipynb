```python
### Write a function that reads the contents of a file named 
# `sample.txt` and prints each line.

def readingFile(file):
    with open(file,'r') as File:
        for line in File:
            print(line.strip())

readingFile("example2.txt")
```

```python
### Write a function that writes a list of strings to a file named 
# `output.txt`, with each string on a new line.

def writingIntoFile(Strings):
    newFile = "output.txt"
    with open(newFile,mode='w',newline="") as File:
        for line in Strings:
            File.write(f"{line} \n")

    with open(newFile,mode='r') as File:
        for line in File:
            print(line)

inputs = ["Python", "File Handling", "Assignments", "Done"]
writingIntoFile(inputs)



```

```python
### Write a function that copies the contents of a file 
# named `source.txt` to a new file named `destination.txt`.

def copying(sourceFile):
    with open(sourceFile,mode='r') as src:
        with open("destination2.txt",mode='w',newline="") as des:
            if des.write(src.read()):
                print(f"The content in source is moved to destination2.txt file")
            else:
                print("There is a mistake in your code")

output = "output.txt"
copying(output)
```

```python
### Write a function that reads the contents of a file named `document.txt` 
# and returns the number of words in the file.

def countWords(file):
    wordCount = 0
    with open(file,mode='r') as File:
        for line in File:
            count = len(line.split())
            wordCount += count
    
    return wordCount

output = "output.txt"
countWords(output)

```

```python
### Write a function that finds and replaces all occurrences of a given word 
# in a file named `data.txt` with another word.

def occurrences(file,word,replaceWord):
    with open(file,mode='r') as File:
        content = File.read()  ### This will convert whole file into a string 
        ## and store in content so we can modify it

        content = content.replace(word,replaceWord)

        with open(file,mode='w',newline="") as File:
            File.write(content)



output = "output.txt"
occurrences(output,"Rahul","Sushma")
```

```python
### Write a function that reads the contents of a file named 
# `reverse.txt` and prints each line in reverse order.

def reverse(file):
    with open(file,mode='r') as File:
        content = File.readlines() ### This will return list of lines , returns list

        for line in content[::-1]:
            print(line)

output = "output.txt"
reverse(output)
```

```python
###Write a function that reads the contents of a file named `stats.txt` 
# and returns the number of lines, words, and characters in the file.

def Counting(file):
    with open(file,mode='r') as File:
        lineCount = len(File.readlines())
        File.seek(0)
        wordCount = len(File.read().split())
        File.seek(0)
        charCount = len(File.read())

        return lineCount,wordCount,charCount
    
output = "output.txt"
Counting(output)  

```

```python
### Write a function that merges the contents of multiple files 
# into a single file named `merged.txt`.

def merging(file1,file2,mergeFile):
    with open(file1,mode='r') as File:
        content1 = File.read()
    with open(file2,mode='r') as File:
        content2 = File.read()
    with open(mergeFile,mode='w') as File:
        File.write(content1 + "\n")
        File.write(content2 + "\n")
    
merging("example.txt","example2.txt","output.txt")

```

```python
### Write a function that reads a CSV file named 
# `data.csv` and prints its contents as a list of dictionaries.

import csv

def CSV(file):
    with open(file,mode='r',newline="") as File:
        read = csv.reader(File)
        output = list(read)
        print(output)

CSV("Data.csv")
```

```python
### Write a function that reads a JSON file named 
# `data.json` and prints its contents as a Python dictionary.

import json

def jsonFile(file):
    with open(file,mode='r',newline="") as File:
        reader = json.load(File)
    
    return reader

value = "data.json"
jsonFile(value)
```
