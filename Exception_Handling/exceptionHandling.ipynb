```python
###Python's exception handling uses try, except, else, and finally blocks to 
# manage runtime errors, prevent program crashes, and handle unexpected 
# situations gracefully. This allows code to detect problems, respond to them, 
# and continue execution.
```

```python
try:
    a=b
except:
    print("The variable is not assigned")
```

```python
### using exception names

try:
    x = 10/0
    print(x)
except ZeroDivisionError as error:
    print(error)
    print(f"Its an ZeroDivision Error")
```

```python
### Base class of exception is Exception
## It can handle any exception

try:
    a=b
except Exception as ex:
    print(ex)
```

```python
### Example with multipul exceptions

try:
    num = int(input("Enter a number"))
    result = 10/num
    print(result)
except ValueError:
    print("Enter only Integers")
except ZeroDivisionError:
    print("Enter values which are greater than 0")
except Exception as ex:
    print(ex)
```

```python
### Try, Except and else block

try:
    num = int(input("Enter a number"))
    result = 10/num
except ValueError:
    print("Enter only Integers")
except ZeroDivisionError:
    print("Enter values which are greater than 0")
except Exception as ex:
    print(ex)
else:
    print(result)
```

```python
### Try, except, else and final block
## final block makes sure no matter what run my block

try:
    num = int(input("Enter a number"))
    result = 10/num
except ValueError:
    print("Enter only Integers")
except ZeroDivisionError:
    print("Enter values which are greater than 0")
except Exception as ex:
    print(ex)
else:
    print(result)
finally:
    print("The program is done")
```

```python
### File handling and exception handling

try:
    file = open('example3.txt',mode='r')
    content = file.read()
except FileNotFoundError:
    print("The is no such file")
else:
    print(content)
finally:
    if 'file' in locals() or not file.closed():
        file.close()
        print("file has been closed")
```
