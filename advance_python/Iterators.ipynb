```python
### An iterator is an object that allows you to traverse through a collection 
# one element at a time and keeps track of its current position. It implements 
# the __iter__() and __next__() methods, and raises StopIteration when no more 
# elements are available.
```

```python
my_list = [1,2,3,4,5,6]

for i in my_list:
    print(f"{i}")   ### this is normal for loop to access all the elements in the list

```

```python
### Now lets use iterator

my_list2 = [7,8,9,10,11,12]

iterator = iter(my_list2)  ### to convert a list into iterator we use iter keyword

print(type(iterator))  ### its a class

### now if you want to iterate through all the elemets use next() keyword

next(iterator)  ## this will output the first value
next(iterator)

```

```python
try:
    print(next(iterator))
except StopIteration:
    print("There are no elements in the iterator")
```

```python
### iterator with string example

my_String = "Rahul"

string_Iterator = iter(my_String)

print(next(string_Iterator))  #### This will give you R
print(next(string_Iterator))  #### This will give you a
```
