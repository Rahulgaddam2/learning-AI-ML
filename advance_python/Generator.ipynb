```python
### A generator is a special type of iterator that generates values one at a 
# time using the yield keyword instead of returning all values at once.

### Generator = An easier way to CREATE an iterator

### The yield keyword is what makes a function a generator.

### yield is ONLY allowed inside a function
```

```python
def largeCount(bigNumber):

    for i in range(bigNumber):  ### range(start, stop, step)
        yield i

genarator = largeCount(100000000)  ### this will not print all 1000000 numbers

print(next(genarator))   ## output 0
print(next(genarator))   ## output 1
print(next(genarator))   ## output 2
print(next(genarator))   ## output 3
```

```python
### Generators with strings

def names():
    yield "rahul"
    yield "sushma"
    yield "adhvay"

Generator = names()

print(next(Generator))  ## output rahul
print(next(Generator))  ## output sushma
print(next(Generator))  ## output adhvay
```
